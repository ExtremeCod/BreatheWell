```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read and parse the data from file 'a1.txt' (breathing signal a)
with open('normal_breath.txt', 'r') as file:
    data_a = file.readlines()
breathing_signal_a = [float(line.strip()) for line in data_a]

# Step 2: Read and parse the data from file 'b.txt' (breathing signal b)
with open('breath_hold.txt', 'r') as file:
    data_b = file.readlines()
breathing_signal_b = [float(line.strip()) for line in data_b]

# Step 3: Normalize the signals to range [0, 1]
def normalize(signal):
    return (signal - np.min(signal)) / (np.max(signal) - np.min(signal))

# Step 4: Apply a simple moving average filter for smoothing
def smooth(signal, window_size=5):
    return np.convolve(signal, np.ones(window_size)/window_size, mode='same')

# Normalize and smooth the signals
breathing_signal_a_normalized = normalize(np.array(breathing_signal_a))
breathing_signal_b_normalized = normalize(np.array(breathing_signal_b))

breathing_signal_a_smoothed = smooth(breathing_signal_a_normalized)
breathing_signal_b_smoothed = smooth(breathing_signal_b_normalized)

# Step 5: Calculate RMSSD (Root Mean Square of Successive Differences)
def rmssd(signal):
    differences = np.diff(signal)
    return np.sqrt(np.mean(differences**2))

rmssd_a = rmssd(breathing_signal_a_smoothed)
rmssd_b = rmssd(breathing_signal_b_smoothed)

# Step 6: Calculate Coefficient of Variation (CV)
def coefficient_of_variation(signal):
    return np.std(signal) / np.mean(signal)

cv_a = coefficient_of_variation(breathing_signal_a_smoothed)
cv_b = coefficient_of_variation(breathing_signal_b_smoothed)

# Step 7: Print the irregularity metrics
print(f"Breathing Signal A - RMSSD: {rmssd_a:.4f}, CV: {cv_a:.4f}")
print(f"Breathing Signal B - RMSSD: {rmssd_b:.4f}, CV: {cv_b:.4f}")

# Step 8: Apply Fourier Transform (FFT) to both smoothed signals
fourier_transform_a = np.fft.fft(breathing_signal_a_smoothed)
fourier_transform_b = np.fft.fft(breathing_signal_b_smoothed)

# Step 9: Calculate the frequency components
frequencies_a = np.fft.fftfreq(len(breathing_signal_a_smoothed))
frequencies_b = np.fft.fftfreq(len(breathing_signal_b_smoothed))

# Step 10: Create a figure with subplots for the time-domain and frequency-domain plots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot smoothed and normalized breathing signal a (Time Domain)
axs[0, 0].plot(breathing_signal_a_smoothed, 'r', 1, linewidth=3.2)
#axs[0, 0].set_title('Smoothed & Normalized Breathing Signal A (Time Domain)')
axs[0, 0].set_xlabel('samples')
axs[0, 0].set_ylabel('Amplitude')
axs[0, 0].grid(True)

# Plot FFT of signal a (Frequency Domain)
axs[0, 1].plot(frequencies_a, np.abs(fourier_transform_a))
#axs[0, 1].set_title('FFT of Smoothed Signal A (Frequency Domain)')
axs[0, 1].set_xlabel('Frequency (Hz)')
axs[0, 1].set_ylabel('Magnitude')
axs[0, 1].grid(True)
axs[0, 1].set_xlim([-0.05, 0.05])
axs[0, 1].set_ylim([0, 200])

# Plot smoothed and normalized breathing signal b (Time Domain)
axs[1, 0].plot(breathing_signal_b_smoothed, 'g', 1, linewidth=3.2)
#axs[1, 0].set_title('Smoothed & Normalized Breathing Signal B (Time Domain)')
axs[1, 0].set_xlabel('samples')
axs[1, 0].set_ylabel('Amplitude')
axs[1, 0].grid(True)

# Plot FFT of signal b (Frequency Domain)
axs[1, 1].plot(frequencies_b, np.abs(fourier_transform_b))
axs[1, 1].set_title('FFT of Smoothed Signal B (Frequency Domain)')
axs[1, 1].set_xlabel('Frequency (Hz)')
axs[1, 1].set_ylabel('Magnitude')
axs[1, 1].grid(True)
axs[1, 1].set_xlim([-0.05, 0.05])
axs[1, 1].set_ylim([0, 200])

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
